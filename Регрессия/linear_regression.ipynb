```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn.linear_model as skl
```

```python
df = pd.read_csv('c-data-1.csv')
df.describe()
np_array = df.to_numpy()

plt.figure(figsize=(12, 6))
plt.scatter(df['x1'], df['y'])
plt.xlabel('x1')
plt.ylabel('y')
plt.grid(True, alpha=0.3)
plt.xlim(0, 80)
plt.ylim(0, 400)
plt.show()
```

```python
x = df[['x1']]
y = df['y']

model_false = skl.LinearRegression(fit_intercept=False)
model_false.fit(x, y)
r2_false = model_false.score(x, y)
print(r2_false)
print(model_false.coef_)
print(model_false.intercept_)

model_true = skl.LinearRegression(fit_intercept=True)
model_true.fit(x, y)
r2_true = model_true.score(x, y)
print(r2_true)
print(model_true.coef_)
print(model_true.intercept_)
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(df['x1'], df['y'], 'bo', alpha=0.7, markersize=4, label='Исходные данные')
plt.plot(df['x1'], model_false.predict(x), 'r-', label='Прогноз (без intercept)')
plt.xlim(df['x1'].min()-5, df['x1'].max()+5)
plt.ylim(df['y'].min()-20, df['y'].max()+20)
plt.grid(True, alpha=0.3)
plt.xlabel('x1')
plt.ylabel('y')
plt.title('Модель без свободного члена (fit_intercept=False)')
plt.legend()


plt.subplot(1, 2, 2)
plt.plot(df['x1'], df['y'], 'bo', alpha=0.7, markersize=4, label='Исходные данные')
plt.plot(df['x1'], model_true.predict(x), 'g-', label='Прогноз (с intercept)')
plt.xlim(df['x1'].min()-5, df['x1'].max()+5)
plt.ylim(df['y'].min()-20, df['y'].max()+20)
plt.grid(True, alpha=0.3)
plt.xlabel('x1')
plt.ylabel('y')
plt.title('Модель со свободным членом (fit_intercept=True)')
plt.legend()

plt.tight_layout()
plt.show()
```
